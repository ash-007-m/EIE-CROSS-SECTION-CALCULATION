```python
import matplotlib.pyplot as plt
import numpy as np

# Read the contents of the file
mapping ={ 1:'3[p-]', 2:'3[p+]',3:'3[d-]',4:'3d[+]',5:'4s',6:'4[p-]',7:'4[p+]',8:'4[d-]',9:'4[d+]',10:'5s',11:'5[p-]',12:'5[p+]',13:'5[d-]',14:'5[d+]',15:'6s',16:'6[p-]',17:'6[p+]',18:'6[d-]',19:'6[d+]',20:'7s',21:'7[p-]',22:'7[p+]',23:'7[d-]',24:'7[d+]'}

def CE_plot(file_path,x,y,up_level):
    with open(file_path, 'r') as file:
     lines = file.readlines()

# Extracting data from the file, skipping the first line
    data = [line.strip().split() for line in lines[100:]]

# Extracting the first and second columns for plotting
    first_column = []
    second_column = []
    for row in data:
        if len(row) >= 4:  # Check if the row has at least two elements
            first_column.append(float(row[x]))
            second_column.append(float(row[y]))

# Plotting
    
    first_column=np.log(first_column)
    second_column=np.log(second_column)
    interval= 100
    first_column= first_column[::interval]
    second_column=second_column[::interval]
    plt.scatter(first_column, second_column, color='red', alpha=0.8, s=50, marker='*')
    plt.plot(first_column, second_column, linestyle='--', color='green')
# Add labels and title
    plt.xlabel('Incident Energy', fontsize=12)
    plt.ylabel('Cross Section', fontsize=12)
    plt.title('3s_' + mapping[up_lev], fontsize=14)

    plt.grid(True)
    plt.show()

for up_lev in range(1,25):
 file_name='CE_CROSS_3s_'+mapping[up_lev]
 CE_plot(file_name,0,2,up_lev)
 

```
